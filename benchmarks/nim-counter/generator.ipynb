```python
for problem_num in range(30):
    with open(f"p1_{problem_num + 1}.pddl", "w") as f:
        f.write("(define (problem nim)\n")
        f.write("\t(:domain nim)\n\n")

        f.write("\t(:objects\n")
        for stone in range(problem_num+1):
            f.write(f"\t\ts1_{stone} - stone\n")
        f.write("\t)\n\n")

        f.write("\t(:init\n")
        f.write(f"\t\t(successor pile1 s1_0)\n")
        for stone in range(1, problem_num+1):
            f.write(f"\t\t(successor s1_{stone - 1} s1_{stone})\n")
        f.write(f"\t\t(successor s1_{problem_num} terminal)\n")
        f.write(f"\t\t(turn p0)\n")
        f.write("\t)\n\n")

        f.write(f"\t(:goal (and\n")
        f.write(f"\t\t(successor pile1 terminal)\n")
        f.write(f"\t\t(opponents_turn)\n")
        f.write("\t))\n")
        f.write(")") 
```

```python
for problem_num in range(10):
    for problem_num2 in range(problem_num, 10):
        with open(f"p2_{problem_num + 1}-{problem_num2 + 1}.pddl", "w") as f:
            f.write("(define (problem nim)\n")
            f.write("\t(:domain nim)\n\n")

            f.write("\t(:objects\n")
            for stone in range(problem_num+1):
                f.write(f"\t\ts1_{stone} - stone\n")
            for stone in range(problem_num2+1):
                f.write(f"\t\ts2_{stone} - stone\n")
            f.write("\t)\n\n")

            f.write("\t(:init\n")
            f.write(f"\t\t(successor pile1 s1_0)\n")
            for stone in range(1, problem_num+1):
                f.write(f"\t\t(successor s1_{stone - 1} s1_{stone})\n")
            f.write(f"\t\t(successor s1_{problem_num} terminal)\n")
            f.write(f"\t\t(successor pile2 s2_0)\n")
            for stone in range(1, problem_num2+1):
                f.write(f"\t\t(successor s2_{stone - 1} s2_{stone})\n")
            f.write(f"\t\t(successor s2_{problem_num2} terminal)\n")
            f.write(f"\t\t(turn p0)\n")
            f.write("\t)\n\n")

            f.write(f"\t(:goal (and\n")
            f.write(f"\t\t(successor pile1 terminal)\n")
            f.write(f"\t\t(successor pile2 terminal)\n")
            f.write(f"\t\t(opponents_turn)\n")
            f.write("\t))\n")
            f.write(")") 
```
