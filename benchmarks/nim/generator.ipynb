```python
for problem_num in range(40):
    with open(f"p1_{problem_num + 1}.pddl", "w") as f:
        f.write("(define (problem nim)\n")
        f.write("\t(:domain nim)\n\n")

        f.write("\t(:objects\n")
        for stone in range(problem_num+1):
            f.write(f"\t\ts{stone} - stone\n")
        f.write(f"\t\tpile1 - pile\n")
        f.write("\t)\n\n")

        f.write("\t(:init\n")
        for stone in range(problem_num+1):
            f.write(f"\t\t(in s{stone} pile1)\n")
        f.write(f"\t\t(turn p0)\n")
        f.write("\t)\n\n")

        f.write(f"\t(:goal (and\n")
        for stone in range(problem_num+1):
            f.write(f"\t\t(not (in s{stone} pile1))\n")
        f.write(f"\t\t(opponents_turn)\n")
        f.write("\t))\n")
        f.write(")") 
```

```python
for problem_num in range(10):
    for problem_num2 in range(10):
        with open(f"p2_{problem_num + 1}-{problem_num2 + 1}.pddl", "w") as f:
            f.write("(define (problem nim)\n")
            f.write("\t(:domain nim)\n\n")

            f.write("\t(:objects\n")
            for stone in range(problem_num+1):
                f.write(f"\t\ts1_{stone} - stone\n")
            for stone in range(problem_num2+1):
                f.write(f"\t\ts2_{stone} - stone\n")
            f.write(f"\t\tpile1 - pile\n")
            f.write(f"\t\tpile2 - pile\n")
            f.write("\t)\n\n")

            f.write("\t(:init\n")
            for stone in range(problem_num+1):
                f.write(f"\t\t(in s1_{stone} pile1)\n")
            for stone in range(problem_num2+1):
                f.write(f"\t\t(in s2_{stone} pile2)\n")
            f.write(f"\t\t(turn p0)\n")
            f.write("\t)\n\n")

            f.write(f"\t(:goal (and\n")
            for stone in range(problem_num+1):
                f.write(f"\t\t(not (in s1_{stone} pile1))\n")
            for stone in range(problem_num2+1):
                f.write(f"\t\t(not (in s2_{stone} pile2))\n")
            f.write(f"\t\t(opponents_turn)\n")
            f.write("\t))\n")
            f.write(")") 
```

```python
for problem_num in range(5):
    for problem_num2 in range(5):
        for problem_num3 in range(5):
            with open(f"p3_{problem_num + 1}-{problem_num2 + 1}-{problem_num3 + 1}.pddl", "w") as f:
                f.write("(define (problem nim)\n")
                f.write("\t(:domain nim)\n\n")

                f.write("\t(:objects\n")
                for stone in range(problem_num+1):
                    f.write(f"\t\ts1_{stone} - stone\n")
                for stone in range(problem_num2+1):
                    f.write(f"\t\ts2_{stone} - stone\n")
                for stone in range(problem_num3+1):
                    f.write(f"\t\ts3_{stone} - stone\n")
                f.write(f"\t\tpile1 - pile\n")
                f.write(f"\t\tpile2 - pile\n")
                f.write(f"\t\tpile3 - pile\n")
                f.write("\t)\n\n")

                f.write("\t(:init\n")
                for stone in range(problem_num+1):
                    f.write(f"\t\t(in s1_{stone} pile1)\n")
                for stone in range(problem_num2+1):
                    f.write(f"\t\t(in s2_{stone} pile2)\n")
                for stone in range(problem_num3+1):
                    f.write(f"\t\t(in s3_{stone} pile3)\n")
                f.write(f"\t\t(turn p0)\n")
                f.write("\t)\n\n")

                f.write(f"\t(:goal (and\n")
                for stone in range(problem_num+1):
                    f.write(f"\t\t(not (in s1_{stone} pile1))\n")
                for stone in range(problem_num2+1):
                    f.write(f"\t\t(not (in s2_{stone} pile2))\n")
                for stone in range(problem_num3+1):
                    f.write(f"\t\t(not (in s3_{stone} pile3))\n")
                f.write(f"\t\t(opponents_turn)\n")
                f.write("\t))\n")
                f.write(")") 
```
